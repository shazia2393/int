```python
# Step 1: Create sample time series data with anomalies Using Z-score for abnormality in timeseies
#Using random function and numpy library from pandas.
np.random.seed(42)
dates = pd.date_range(start='01-01-2017', periods=1000, freq='5T')  # 5-minute intervals
values = np.random.normal(loc=50, scale=10, size=1000)  # Normal data around 50
```

```python
# Introduce some anomalies (extreme values)
values[100:110] = [150, 140, 160, 130, 155, 145, 150, 148, 160, 162]  # Abnormal spike
```

```python
# Create DataFrame
del_df = pd.DataFrame({'time': dates, 'Cyclone_Inlet_Gas_Temp': values})
del_df.set_index('time', inplace=True)

```

```python
# Step 2: Z-Score based anomaly detection
del_df['Z-Score'] = (del_df['Cyclone_Inlet_Gas_Temp'] - del_df['Cyclone_Inlet_Gas_Temp'].mean()) / del_df['Cyclone_Inlet_Gas_Temp'].std()
```

```python
# Mark anomalies where Z-Score > 3 or Z-Score < -3
del_df['Z-Anomaly'] = (del_df['Z-Score'].abs() > 3).astype(int)
```

```python
# Step 3: Isolation Forest for anomaly detection
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01)  # 1% of data expected as anomalies
del_df['IF_Anomaly'] = iso_forest.fit_predict(del_df[['Cyclone_Inlet_Gas_Temp']])

```
